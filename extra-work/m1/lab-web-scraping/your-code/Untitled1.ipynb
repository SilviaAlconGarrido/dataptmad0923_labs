```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import bs4
import re
```

```python
# This is the url you will scrape in this exercise
url = 'https://amazonasmascotas.com/20-pienso-para-perros'
```

```python
response =requests.get(url) # llamo a la api
#type(response)
```

```python
html = response.content #contenido

len(html)#longitud del contenido
```

```python
parsed_html = bs4.BeautifulSoup(html, 'html.parser')#parseo

#parsed_html.head
#parsed_html.body
#parsed_html.title

parsed_html.body.a
```

```python
article = parsed_html.find_all('div',{'class':'pro_second_box pro_block_align_1'})
article
```

```python
data = []
for product in article:
    title = product.find('h3', {'class': 's_title_block flex_child'}).text.strip()
    price = product.find('div', {'class': 'product-price-and-shipping pad_b6'}).text.strip()
    description = product.find('div', {'class': 'product-desc pad_b6'}).text.strip()
    data.append([title, price, description])

    df = pd.DataFrame(data, columns=['Título', 'Precio', 'Descripción'])
df
```
