```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import bs4
import re
```

```python
# This is the url you will scrape in this exercise
url = 'https://www.kiwoko.com/perros/comida-para-perros/pienso-seco-para-perros/'
```

```python
response =requests.get(url) # llamo a la api
#type(response)
```

```python
html = response.content #contenido

len(html)#longitud del contenido
```

```python
parsed_html = bs4.BeautifulSoup(html, 'html.parser')#parseo

#parsed_html.head
#parsed_html.body
#parsed_html.title

parsed_html.body.a
```

```python
article = parsed_html.find_all('h2',{'class':'text-base js-product-tile-name gtm-product-name'})
```

```python
len(article)
```

```python
article_list = [] #creo una lista vacia
```

```python
for h2 in article:
    h2 = str(h2)
    soup_a =bs4.BeautifulSoup(h2,'html.parser')
    #print(soup_a)
    list_a = soup_a.find_all('h2')
    #print(list_a)
    for e in list_a:
        article_list.append(e.string.strip())
print(len(article_list))

article_list
```

```python
#find 1 'h1' tag:
#article = parsed_html.find('h1')
#find_all 'h1' tag:
article_price = parsed_html.find_all('span',{'class':'value js-product-tile-price'})
article_price
```

```python
article_list_price = [] #creo una lista vacia
```

```python
for span in article:
    span = str(span)
    soup_a =bs4.BeautifulSoup(span,'html.parser')
    #print(soup_a)
    list_a = soup_a.find_all('span')
    #print(list_a)
    for e in list_a:
        article_list_price.append(e.string.strip())
print(len(article_list_price))

article_list_price
```
